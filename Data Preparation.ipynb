### 2: Load The PLOD Dataset

```python
from datasets import load_dataset, load_metric
data = load_dataset("surrey-nlp/PLOD-CW")
```

Function to convert dictionary into csv format

```python
import csv

# Function to convert the dataset to CSV format
def convert_to_csv(dataset, split_name, output_filename):
    # Open the CSV file for writing
    with open(output_filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)

        # Write the header row
        writer.writerow(['index', 'Sentence #', 'Word', 'POS', 'Tag'])

        # Initialize the sentence counter
        sentence_counter = 1

        # Iterate through each sentence in the dataset
        for idx, sentence in enumerate(dataset[split_name]):
            tokens = sentence['tokens']
            pos_tags = sentence['pos_tags']
            ner_tags = sentence['ner_tags']

            # Write each token in the sentence
            for token_index in range(len(tokens)):
                # If it's the first token in the sentence, indicate the sentence number
                if token_index == 0:
                    writer.writerow([idx, f'Sentence: {sentence_counter}', tokens[token_index], pos_tags[token_index], ner_tags[token_index]])
                else:
                    writer.writerow([idx, 'NaN', tokens[token_index], pos_tags[token_index], ner_tags[token_index]])

            # Increment the sentence counter
            sentence_counter += 1
```

Convert All Datasets to csv format

```python
# Convert train, validation, and test datasets to CSV files
convert_to_csv(data, 'train', 'train_data.csv')
convert_to_csv(data, 'validation', 'validation_data.csv')
convert_to_csv(data, 'test', 'test_data.csv')
```

Concatenate the newly obtained csv files into a single file

```python
import pandas as pd

# Read the datasets
train = pd.read_csv("train_data.csv", encoding="latin1")
test = pd.read_csv("test_data.csv", encoding="latin1")
validate = pd.read_csv("validation_data.csv", encoding="latin1")

# Combine the datasets using pd.concat
train_data = pd.concat([train, test, validate])

# Reset the index of the combined dataset (optional)
train_data.reset_index(drop=True, inplace=True)

# Save the combined dataset to a new CSV file (optional)
train_data.to_csv("combined_train_data.csv", index=False, encoding="latin1")
```
